## Nội dung bài
Hiển thị ngẫu nhiên $N=50$ hình tròn:
+ Tâm $c_i = (x_i, y_i)$ ngẫu nhiên với $x_i, y_i \in [0, 1]$
+ Màu sắc $color_i$ ngẫu nhiên với $color_i \in [0, 1]$
+ Diện tích $area_i = \pi * r_i^2$ với sinh ngẫu nhiên $r_i \in [0, 1]$

## Bài làm

### Import thư viện

```python
import numpy as np
import matplotlib.pyplot as plt
from IPython import display  # hiển thị rich output
```

### Sinh ngẫu nhiên hình tròn

```python
%time
# generate randomly the center (x,y) of circle, color, area
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (15 * np.random.rand(N)) ** 2  # bán kính từ 0 tới 15
```

```python
# Mã kiểm tra dữ liệu
print(f'x: len={len(x):d}, [min={min(x):.2f}, [max={max(x):.2f}]]: 10 samples [{x[0:10]}]')
print(f'y: len={len(y):d}, [min={min(y):.2f}, [max={max(y):.2f}]]: 10 samples [{y[0:10]}]')
print(f'colors: len={len(colors):d}, [min={min(colors):.2f}, [max={max(colors):.2f}]]: 10 samples [{colors[0:10]}]')
print(f'area: len={len(area):d}, [min={min(area):.2f}, [max={max(area):.2f}]]: 10 samples [{area[0:10]}]')
```

### Hiển thị đồ thị

```python
%matplotlib inline
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```
